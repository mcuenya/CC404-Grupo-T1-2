```python
# 0. Crear columna respondieron (1 = respondió, 0 = no respondió)
df_final["respondieron"] = np.where(df_final["ITF"].notna() & (df_final["ITF"] >= 0), 1, 0)

# 1. Cargar tabla equivalencias
tabla_adulto_equiv_csv = """
Edad,Mujeres,Varones
Menor de 1 año,0.35,0.35
1 año,0.37,0.37
2 años,0.46,0.46
3 años,0.51,0.51
4 años,0.55,0.55
5 años,0.6,0.6
6 años,0.64,0.64
7 años,0.66,0.66
8 años,0.68,0.68
9 años,0.69,0.69
10 años,0.7,0.79
11 años,0.72,0.82
12 años,0.74,0.85
13 años,0.76,0.9
14 años,0.76,0.96
15 años,0.77,1
16 años,0.77,1.03
17 años,0.77,1.04
18 a 29 años,0.76,1.02
30 a 45 años,0.77,1
46 a 60 años,0.76,1
61 a 75 años,0.67,0.83
más de 75 años,0.63,0.74
"""
tabla_equiv = pd.read_csv(StringIO(tabla_adulto_equiv_csv))

# 2. Función para asignar equivalencia según edad y sexo
def obtener_equivalencia(edad, sexo):
    try:
        if edad < 1:
            fila = tabla_equiv.iloc[0]
        elif edad == 1:
            fila = tabla_equiv[tabla_equiv["Edad"]=="1 año"].iloc[0]
        elif edad == 2:
            fila = tabla_equiv[tabla_equiv["Edad"]=="2 años"].iloc[0]
        elif edad == 3:
            fila = tabla_equiv[tabla_equiv["Edad"]=="3 años"].iloc[0]
        elif edad == 4:
            fila = tabla_equiv[tabla_equiv["Edad"]=="4 años"].iloc[0]
        elif edad == 5:
            fila = tabla_equiv[tabla_equiv["Edad"]=="5 años"].iloc[0]
        elif edad == 6:
            fila = tabla_equiv[tabla_equiv["Edad"]=="6 años"].iloc[0]
        elif edad == 7:
            fila = tabla_equiv[tabla_equiv["Edad"]=="7 años"].iloc[0]
        elif edad == 8:
            fila = tabla_equiv[tabla_equiv["Edad"]=="8 años"].iloc[0]
        elif edad == 9:
            fila = tabla_equiv[tabla_equiv["Edad"]=="9 años"].iloc[0]
        elif edad == 10:
            fila = tabla_equiv[tabla_equiv["Edad"]=="10 años"].iloc[0]
        elif edad == 11:
            fila = tabla_equiv[tabla_equiv["Edad"]=="11 años"].iloc[0]
        elif edad == 12:
            fila = tabla_equiv[tabla_equiv["Edad"]=="12 años"].iloc[0]
        elif edad == 13:
            fila = tabla_equiv[tabla_equiv["Edad"]=="13 años"].iloc[0]
        elif edad == 14:
            fila = tabla_equiv[tabla_equiv["Edad"]=="14 años"].iloc[0]
        elif edad == 15:
            fila = tabla_equiv[tabla_equiv["Edad"]=="15 años"].iloc[0]
        elif edad == 16:
            fila = tabla_equiv[tabla_equiv["Edad"]=="16 años"].iloc[0]
        elif edad == 17:
            fila = tabla_equiv[tabla_equiv["Edad"]=="17 años"].iloc[0]
        elif 18 <= edad <= 29:
            fila = tabla_equiv[tabla_equiv["Edad"]=="18 a 29 años"].iloc[0]
        elif 30 <= edad <= 45:
            fila = tabla_equiv[tabla_equiv["Edad"]=="30 a 45 años"].iloc[0]
        elif 46 <= edad <= 60:
            fila = tabla_equiv[tabla_equiv["Edad"]=="46 a 60 años"].iloc[0]
        elif 61 <= edad <= 75:
            fila = tabla_equiv[tabla_equiv["Edad"]=="61 a 75 años"].iloc[0]
        else:
            fila = tabla_equiv[tabla_equiv["Edad"]=="más de 75 años"].iloc[0]

        return fila["Varones"] if sexo==1 else fila["Mujeres"]
    except:
        return np.nan

# 3. Crear adulto_equiv solo para respondieron
df_final["adulto_equiv"] = np.where(
    df_final["respondieron"]==1,
    df_final.apply(lambda x: obtener_equivalencia(x["CH06"], x["CH04"]), axis=1),
    np.nan
)

# 4. Calcular adulto_equiv por hogar
df_final["ad_equiv_hogar"] = df_final.groupby("CODUSU")["adulto_equiv"].transform("sum")

# 5. Calcular ingreso necesario solo para respondieron
def ingreso_necesario(row):
    if row["respondieron"]==0:
        return np.nan
    if row["ANO4"] == 2005:
        return row["ad_equiv_hogar"] * 205.07
    elif row["ANO4"] == 2025:
        return row["ad_equiv_hogar"] * 365.177
    else:
        return np.nan

df_final["ingreso_necesario"] = df_final.apply(ingreso_necesario, axis=1)

# 6. Columna pobre solo para respondieron
df_final["pobre"] = np.where(
    df_final["respondieron"]==1,
    (df_final["ITF"] < df_final["ingreso_necesario"]).astype(int),
    np.nan
)
```

```python
# Para este caso 
# Reemplazar valores de CH14 según diccionario
dic_ch14 = {
    'Ninguno': 0,
    'Primero': 1,
    'Segundo': 2,
    'Tercero': 3,
    'Cuarto': 4,
    'Quinto': 5,
    'Sexto': 6,
    'Séptimo': 7,
    'Octavo': 8,
    'Noveno': 9,
    'Educación especial': np.nan  # dejamos como NaN, porque el numero de las referencias es 98 pero es muy incoherente decir que tienen 98 años de educacion
}

df_final['CH14'] = df_final['CH14'].replace(dic_ch14)

# 4. Función para calcular años de educación
def calcular_educ(row):
    nivel = row["CH12"]   # Nivel educativo (1 a 9)
    completo = row["CH13"]  # 1 = Sí, 2 = No
    anio = row["CH14"]    # Último año aprobado (0 a 9)

    if pd.isna(nivel) or pd.isna(completo) or pd.isna(anio):
        return np.nan

    # Jardín / Preescolar
    if nivel == 1:
        return 0  

    # Primario o EGB (7 años)
    elif nivel in [2, 3]:
        return 7 if completo == 1 else anio

    # Secundario o Polimodal (12 años)
    elif nivel in [4, 5]:
        return 12 if completo == 1 else 7 + anio

    # Terciario (15 años aprox.)
    elif nivel == 6:
        return 15 if completo == 1 else 12 + anio

    # Universitario (17 años aprox.)
    elif nivel == 7:
        return 17 if completo == 1 else 12 + anio

    # Posgrado
    elif nivel == 8:
        return 19

    # Educación especial
    elif nivel == 9:
        return np.nan

    return np.nan

# 5. Crear variable educ
df_final["educ"] = df_final.apply(calcular_educ, axis=1)

# 6. Aplicar tope razonable de años de educación
tope_educ = 25
df_final.loc[df_final["educ"] > tope_educ, "educ"] = np.nan

# 7. Estadísticas descriptivas
stats = df_final["educ"].describe(percentiles=[0.5]).round(2)
print(stats)
```
